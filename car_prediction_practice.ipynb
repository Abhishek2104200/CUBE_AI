```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
df=pd.concat([df,ohe_trans],axis=1)
```
