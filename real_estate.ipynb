```python
df_encoded = df_encoded[['listPrice', 'baths','beds', 'garage', 'sqft', 'lot_sqft', 'type_land',
       'type_mobile', 'type_other', 'type_single_family',
       'status_ready_to_build']]
```

```python
X = df_encoded.drop(['listPrice'], axis=1)
b = df_encoded['listPrice']

# Splitting test and train sets and fitting random forest model
X_train, X_test, y_train, y_test = train_test_split(X, b, test_size=0.2, random_state=42)
rf = RandomForestRegressor()

rf.fit(X_train, y_train)
```

```python
rf.predict(X_test)
```
